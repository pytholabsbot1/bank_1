```python
nom = pd.read_excel(xls,'NOMINATION')
# nom = nom.applymap(str)
```

```python
nom.columns[1:]
```

```python
p = {}
r = 0

errors = {}

def nominations():
    global p ,r 
    cols = nom.columns[1:]
    
    models.client.objects.all().delete()
    for index,row in nom.iterrows():    
        r = row
        profile_data = {c.strip() : row[c] for c in cols if(str(row[c]) not in ['nan','NaT'])}
        
        print(profile_data)
        
        profile_data['id_type'],_ = models.Id_Type_Choice.objects.get_or_create(id_type_choice= str(row['id_type']))
        profile_data['state'],_ = models.State_Choice.objects.get_or_create(state_choice= str(row['state']))
        profile_data['country'],_ = models.Country_Choice.objects.get_or_create(country_choice= str(row['country']))
        profile_data['district'],_ = models.District_Choice.objects.get_or_create(district_choice= str(row['district']))

        profile_data['city'],_ = models.City_Choice.objects.get_or_create(city_choice= str(row['city']))
        profile_data['nominee_relation'],_  = models.Relation_Choice.objects.get_or_create(relation_choice=str(row['nominee_relation'])) 
        
        p = profile_data
        
        try:
            c = models.client(**p)
            c.save()
        except:
            errors[index] = profile_data
            print('eroor '+" ........ >>>>>")

Thread(target=nominations ,).start()  
```

```python
p=r=0 
soc_data = pd.read_excel(xls,'SOCIETY')

def society():
    global p ,r 
    cols = soc_data.columns[1:]
    print(cols)
    
    models.deposits_table.objects.all().delete()
    p_ = models.client.objects.all()
    
    for index,row in soc_data.iterrows():
        r = row
        profile_data = {c.strip() : row[c] for c in cols if(str(row[c]) not in ['nan','NaT','D'])}
        print(profile_data)
        profile_data['scheme'] = models.Scheme.objects.get(name= str(row['scheme']))
        profile_data['person'] = p_[np.random.randint(len(p_))]
        profile_data['category'] = models.DepositChoice.objects.get(name="DD")
        
        
        p = profile_data
        print(profile_data)
        
        try:
            c = models.deposits_table(**p)
            c.save()
        except:
            errors[index] = profile_data
            print('eroor '+" ........ >>>>>")
        
Thread(target=society ,).start()
```

```python
p=r=0 
fc_data = pd.read_excel(xls,'APPROVAL')

def fc_push():
    global p ,r 
    cols = fc_data.columns
    
    models.approved_finance_table.objects.all().delete()
    p_ = models.client.objects.all()
    
    for index,row in fc_data.iterrows():
        r = row
        profile_data = {c.strip() : row[c] for c in cols if(str(row[c]) not in ['nan','NaT','D'])}
        print(profile_data)
        profile_data['person'] = p_[np.random.randint(len(p_))]
        profile_data['emi_type'] = "Daily"
        profile_data['duration_type'] = "days"
        profile_data['roi'] = 10
        
        try:
            profile_data['finance'] = models.finance_table.objects.get(loan_account_number=profile_data['loan_account_number'])
            profile_data['person'] = profile_data['finance'].person
        except:
            errors[index] = profile_data
            continue
            
        if('client_bank' in profile_data.keys()):
            profile_data['client_bank'],_ = models.Bank_Choice.objects.get_or_create(bank_choice=profile_data['client_bank'])
        
        del  profile_data['loan_account_number']
        p = profile_data
        print(profile_data)
        
        try:
            c = models.approved_finance_table(**p)
            c.save()
        except:
            errors[index] = profile_data
            print('eroor '+" ........ >>>>>")
        
Thread(target=fc_push ,).start()
```
