```python
d_max = 100
d_curr = 0

def dp_coll(arg):
    global d_curr
    print('new  -------------------------- ')
    args = chunks(models.deposits_table.objects.all() , 1)[arg]
    
    for deposit in args:
        
#         print(deposit.agent_name.nomination_number)
#         break
        
        start_date = deposit.account_opening_date
        
        if(deposit.category.name=='CA'):
            d = np.random.randint(4,30)
            t_delta = relativedelta(days=+d)
            
        elif(deposit.category.name=='RD'):
            t_delta = relativedelta(months=+1)
            
        else:
            continue
            
        
        end_date = start_date + t_delta
        
        for i in range(np.random.randint(20,50)):

            tot_interest = deposit.interest_collected 
            balance = deposit.balance

            payment_received = np.random.choice(np.arange(500 , 2000 ,100))

            ##Computing --------
            latest_col = deposit.collection_deposit_set.all().order_by('-id')     
            if(latest_col):
                days = abs((end_date - latest_col[0].payment_received_date).days)
            else:
                days = abs((end_date - deposit.account_opening_date).days)

              #calculate interest till date from last collection
            interest_delta = (balance * deposit.scheme.per_day_roi) * days

            #add with last interest 
            tot_interest +=  interest_delta


            ##Updating on database ----------
            k_ = {"deposit" : deposit,
                 "payment_received_date" : end_date ,
                  "payment_received" : payment_received,
                  "previous_balance" : balance,
                  "agent_name" : deposit.agent_name,
                  "latest_intrest" : tot_interest,
                  "person" : deposit.person
                 }


            models.collection_deposit(**k_).save()
            end_date += t_delta

            #update in employee records
            deposit.agent_name.deposit_coll += payment_received
            deposit.agent_name.save()

            deposit.interest_collected += interest_delta
            deposit.balance += payment_received
            deposit.save()

            #update for admin insights in category
            c = models.DepositChoice.objects.get(id=deposit.category.id)
            c.amount += payment_received
            c.save()
            
                
        if(d_curr%20 ==0):
            clear_output()

        d_curr+=1

        p = int( d_curr / d_max *100 )
        print('i'*p , p,"%")

    #         print(payment_received , ' ----- -')

    
for arg in range(10):
# reload(models)
    Thread(target=dp_coll ,args=(arg , )).start()  
```

```python
def fc_coll(arg):
     
    print('new  -------------------------- ')
    finances = chunks(models.approved_finance_table.objects.all() , 1)[arg]
    
    for finance in finances:
        
        try:
            d_ = finance.emi_delta.split()
            print(d_)
            delta = eval(f"relativedelta({d_[1]}=+{d_[0]})")


            emi_date = finance.loan_start_date + delta

            emp = finance.finance.agent_name

            for i in range(finance.no_of_emi):

                k_ = {'finance': finance,
                     "loan_emi_received_date" : emi_date,
                     "agent_name": emp,
                      "person" : finance.finance.person,
                      "loan_emi_received" : finance.emi_amount

                     }


                models.collection_finance(**k_).save()


                 #update in employee records
                emp.finance_coll += finance.emi_amount
                emp.save()

                finance.recieved += finance.emi_amount
                finance.save()

                #update for admin insights in category
                c = models.FinanceChoice.objects.get(id=finance.finance.loan_type.id)
                c.amount += finance.emi_amount
                c.save()
                
        except:
            print('error')
            
        print(finance)

for arg in range(10):
    Thread(target=fc_coll ,args=(arg , )).start()  
```

```python
def pop_cashColl(arg):
    start_date = parser.parse("2015-04-01")
    end_date = start_date
    print('started')
    
    print('new  -------------------------- ')
    args = chunks(models.employee_interview.objects.all() , 5)[arg]
    
    for i ,e in enumerate(args):
        print("employee" ,e.first_name)
        for day in range(400):
            end_date += relativedelta(days=+1)
#             print("end_date ", end_date)

            today_d = e.collection_deposit_set.filter(payment_received_date=end_date).values_list("payment_received")

            today_f = e.collection_finance_set.filter(loan_emi_received_date=end_date).values_list("loan_emi_received")

            day_total = sum([i[0] for i in today_d]) + sum([i[0] for i in today_f])

            if(day_total):
                k_ = {
                    "employee": e,
                    "total_society_deposit" : sum([i[0] for i in today_d]) ,
                    "total_loan_deposit" : sum([i[0] for i in today_f]),
                    "date" : end_date,
                    "remarks" : "-"
                }
                models.cash_collection(**k_).save()

                print(f"Thread:{arg} emp :{i} / {len(args)}  \ncreated on " ,end_date)

        end_date = start_date
        print("resetting" , end_date)

for arg in range(2):
    Thread(target=pop_cashColl ,args=(arg , )).start()  
```

```python
def dp_withdrawl():
    
    for dp in models.deposits_table.objects.all():
        
        if('FD' in dp.scheme.name):
            cat = "Maturity"
            w_date = dp.maturity_date 
            interest = dp.maturity_amount
            w_amount = dp.maturity_amount
            
        else:
            cat = "withdraw"
            w_date = dp.account_opening_date + relativedelta(months=+1)
            interest = dp.scheme.per_day_roi * (w_date - dp.account_opening_date).days * dp.account_opening_amount
            w_amount = np.random.randint(200,2000)
            
        k_ ={
            "society_account" : dp,
            "holder_name" : dp.person.first_name,
            "category" : cat,
            "amount_withdrawl_date" : w_date,
            "available_amount" : dp.balance,
            "intrest_amount" : interest,
            "amount_withdrawl" : w_amount
        }
        
        models.withdrawl_entry(**k_).save()
        print(dp)
        
Thread(target=dp_withdrawl).start() 
```
